Data analysis project with 'Covid-19 Dataset'

```python
import pandas as pd
```

```python
#reading the data from csv file to the pandas dataframe
df = pd.read_csv('covid_19_data.csv')
df.head(5)
```

Q1. Show the number of Confirmed, Deaths and Recovered cases in each Region.

```python
df.groupby('Region').sum('numeric_only')
#we use head() and tail() functions to view top & bottom rows
```

Q2. Remove all the records where Confirmed Cases is Less Than 10.

```python
df = df[~(df.Confirmed < 10)]
```

Q3. In which Region, Maximum number of Confirmed cases were recorded?

```python
df = pd.read_csv('covid_19_data.csv')
```

```python
df.groupby('Region').Confirmed.sum().sort_values(ascending = False)
```

Q4. In which Region, minimum number of Deaths cases were recorded?

```python
df.groupby('Region').Deaths.sum().sort_values().head(30)
```

Q5. How many Confirmed, Deaths & Recovered cases were reported from India?

```python
df[df.Region == 'India']
```

Q6. Sort the entire data wrt No. of Confirmed cases in ascending order.

```python
df.sort_values(by = ['Confirmed'])
```

Q6. Sort the entire data wrt No. of Recovered cases in decending order.

```python
df.sort_values(by = ['Recovered'],ascending = False)
```

Q7. What is the Deaths percentage in India?

```python
a = df[df.Region == 'India']
percentage=(a['Deaths']/a['Confirmed'])*100
print(percentage)
```

Q8. Find the total number of Confirmed cases recorded in the world?

```python
df = pd.read_csv('covid_19_data.csv')
df['Confirmed'].sum()
```

Q9. Find the average number of Recovered cases in the world?

```python
recov = df['Recovered'].sum()
length = len(df)
average = (recov/length)
print(average)
```
